# 3. Merging dataframes

```python
df_orders_products_combined.head()
```

```python
# Drop _merged column in view of coming merge below

df_orders_products_combined.drop(['_merge'], axis = 1, inplace = True)
```

```python
# Drop Unnamed: 0 column in view of coming merge below

df_orders_products_combined.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df_products_checked.head()
```

```python
# Drop Unnamed: 0 column in view of coming merge below

df_products_checked.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df_products_checked.head()
```

```python
# Checking for duplicates in df_orders_products combined

df_dups_opc = df_orders_products_combined[df_orders_products_combined.duplicated()]
```

```python
df_dups_opc.head()
```

No duplicates found.

```python
# Checking for duplicates in products checked

df_dups_pc = df_products_checked[df_products_checked.duplicated()]
```

```python
df_dups_pc.head()
```

No full duplicates found.

```python
# Checking for duplicates in product_id column

df_products_checked['product_id'].value_counts()
```

We found duplicates for product_id 6800 and 26520

```python
df_products_checked.iloc[6782:]
```

```python
df_products_checked.iloc[26503:]
```

We will correct the two duplicate values by assigning the most probable correct product_id in both cases.

```python
# Correcting product_id for 6800 duplicate

df_products_checked.at[6784, 'product_id'] = 6799
```

```python
# Correcting prodcut_id for 26520 duplicate

df_products_checked.at[26504, 'product_id'] = 26519
```

```python
# Test merge

pd.merge(df_orders_products_combined, df_products_checked, on = ['product_id'], how = 'left', indicator = True)
```

Left merge leads to equal number of rows post merge. Proceeding with actual merge below.

```python
# Executing left merge of dataframes

df_orders_products_merged = df_orders_products_combined.merge(df_products_checked, on = ['product_id'], how = 'left', indicator = True)
```

```python
df_orders_products_merged.shape
```

```python
df_orders_products_merged.head()
```

```python
# Confirm results using merge flag

df_orders_products_merged['_merge'].value_counts()
```

```python
# Checking for duplicates in newly merged df

df_dups_merged = df_orders_products_merged[df_orders_products_merged.duplicated()]
```

```python
df_dups_merged.head()
```

No duplicates identified.

```python
# Exporting newly merged dataframe in pickle format

df_orders_products_merged.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```
