### b. Find the aggregated mean of the "order_number" column grouped by "department_id"

### d. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions

```python
# Create a new column containing the maximum frequency of the “order_number” column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create a second column containing a flag designating whether a customer is “loyal” or not

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 40, 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Count customers sorted by flag

ords_prods_merge['loyalty_flag'].value_counts()
```

```python
ords_prods_merge.head()
```

### e. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

The mean price of products bought by the three customer categories is nearly identical. Loyal customers displays the lowest mean product price of the three.

### f. Create a spending flag for each user based on the average price across all their orders

```python
# Determining average price for each user across all their orders

ords_prods_merge.groupby(['user_id']).agg({'prices': ['mean']})
```

```python
# Create a new column "spender_type" containing mean price spent per user_id

ords_prods_merge['spender_type'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Create a spending flag for each customer based on average price 

ords_prods_merge.loc[ords_prods_merge['spender_type'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spender_type'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Displaying result

ords_prods_merge[['user_id', 'spender_type', 'spending_flag']].head(500)
```

### g. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
# Create a new column "median_order_frequency" containing the regularity of user_id spending

ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Create a flag for each customer depending on order frequency 

ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'customer_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] > 10) & (ords_prods_merge['median_order_frequency'] <= 20), 'customer_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'customer_frequency_flag'] = 'Frequent customer'
```

```python
#Displaying result

ords_prods_merge[['user_id', 'median_order_frequency', 'customer_frequency_flag']].head(500)
```
